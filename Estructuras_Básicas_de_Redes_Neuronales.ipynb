# **Representaciones de datos para redes neuronales.**

**Tensores**

Los tensores son objetos matemáticos que almacenan valores numéricos y que pueden tener distintas dimensiones.

Los tensores son fundamentales para este campo, tan fundamentales que TensorFlow recibió su nombre. Un tensor es un contenedor de datos, generalmente datos numéricos. Entonces, es un contenedor de números.

Los tensores son una generalización de matrices a un número arbitrario de dimensiones (tenga en cuenta que en el contexto de los tensores, una dimensión a menudo se denomina eje ).

### Escalares (tensores de rango 0)

Un tensor que contiene solo un número se llama escalar (o tensor escalar, o tensor de rango 0, o tensor 0D). En NumPy, un número float32 o float64 es un tensor escalar (o matriz escalar). **Al número de ejes de un tensor también se le llama rango**


````python
# Escalares (tensores de rango 0)
import numpy as np
escalar = np.array(12)
escalar
````

````python
# Mostrar el número de ejes de un tensor - ndim
escalar.ndim
````

## Vectores (tensores de rango 1)

Una matriz de números se llama vector, tensor de rango 1 o tensor 1D. Se dice que un tensor de rango 1 tiene exactamente un eje.

````python
# Vectores (tensores de rango 1)
vector = np.array([12,3,6,14,7])
vector
````

````python
# Mostrar el número de ejes - ndim
vector.ndim
````

Este vector tiene cinco entradas y por eso se llama vector de 5 dimensiones. ¡No confundir un vector 5D con un tensor 5D! Un vector 5D tiene solo un eje y cinco dimensiones a lo largo de su eje, mientras que un tensor 5D tiene cinco ejes (y puede tener cualquier número de dimensiones a lo largo de cada eje). La dimensionalidad puede denotar ya sea el número de entradas a lo largo de un eje específico (como en el caso de nuestro vector 5D) o el número de ejes en un tensor (como un tensor 5D), lo que a veces puede resultar confuso. En el último caso, es técnicamente más correcto hablar de un tensor de rango 5 (el rango de un tensor es el número de ejes), pero la notación ambigua tensor 5D es común de todos modos.

## Matrices (tensores de rango 2)
Una matriz de vectores es una matriz, un tensor de rango 2 o un tensor 2D. Una matriz tiene dos ejes (a menudo denominados filas y columnas).

````python
# Matrices (tensores de rango 2)
matriz = np.array([[5 , 78 , 2 , 34 , 0],
                  [6 , 79 , 3 , 35 , 1],
                  [7 , 80 , 4 , 36 , 2]])
````

````python
# Mostrar el número de ejes - ndim
matriz.ndim
````

Las entradas del primer eje se llaman filas y las entradas del segundo eje se denominan columnas. En el ejemplo anterior, [5, 78, 2, 34, 0] es la primera fila de x, y [5, 6, 7] es la primera columna.

## Tensores de rango 3 y de rango superior

Si empaqueta dichas matrices en una nueva matriz, obtendrá un tensor de rango 3 (o tensor 3D), que puede interpretar visualmente como un cubo de números.

````python
# Tensores de rango 3 y de rango superior
tensor3 = np.array([[[ 5 , 78 , 2 , 34 , 0 ],
                   [ 6 , 79 , 3 , 35 , 1 ],
                   [ 7 , 80 , 4 , 36 , 2 ]],
                  [[ 5 , 78 , 2 , 34 , 0 ],
                   [ 6 , 79 , 3 , 35 , 1 ],
                   [ 7 , 80 , 4 , 36 , 2 ]],
                  [[ 5 , 78 , 2 , 34 , 0 ],
                   [ 6 , 79 , 3 , 35 , 1 ],
                   [ 7 , 80 , 4 , 36 , 2 ]]])
````

````python
# Mostrar el número de ejes - ndim
tensor3.ndim
````

Al empaquetar tensores de rango 3 en una matriz, puede crear un tensor de rango 4, y así sucesivamente. En el aprendizaje profundo, generalmente manipularás tensores con rangos del 0 al 4, aunque puedes llegar hasta el 5 si procesas datos de video.

# Aspectos claves

Un tensor se define por tres atributos clave:


> **Número de ejes (rango):** por ejemplo, un tensor de rango 3 tiene tres ejes y una matriz tiene dos ejes. Esto también se llama tensor ndim en bibliotecas de Python como NumPy o TensorFlow.


> **Forma (Shape):** es una tupla de números enteros que describe cuántas dimensiones tiene el tensor a lo largo de cada eje. Por ejemplo, el ejemplo de matriz anterior tiene forma (3, 5) y el ejemplo de tensor de rango 3 tiene forma (3, 3, 5). Un vector tiene una forma con un solo elemento, como (5,), mientras que un escalar tiene una forma vacía ().



> **Tipo de datos** (generalmente llamado dtype en las bibliotecas de Python): esto es el tipo de datos contenidos en el tensor; por ejemplo, el tipo de tensor podría ser float16, float32, float64, uint8, etc. En TensorFlow, también es probable que te encuentres con tensores string.









# Ejemplos de tensores de datos



> **Datos vectoriales:** tensores de rango 2 de forma (samples, features), donde cada muestra es un vector de atributos numéricos (“características”)

> **Datos de series temporales o datos de secuencia:** tensores de rango 3 de forma (samples, timesteps, features), donde cada muestra es una secuencia (de longitud timesteps) de vectores de características.

> **Imágenes:** tensores de forma de rango 4 (samples, height, width, channels), donde cada muestra es una cuadrícula 2D de píxeles y cada píxel está representado por un vector de valores ("canales")


> **Vídeo:** tensores de forma de rango 5 (samples, frames, height, width, channels), donde cada muestra es una secuencia (de longitud frames) de imágenes.

````python
#Importamos la librería
import numpy as np
````

````python
# Creamos funciones

# Definición e implementación de las funciones necesarias
############################################
def sigmoide(sdp):
    return 1.0/(1+np.exp(-1*sdp))

def error(predicho, target):
    return np.power(predicho-target,2) #Primeros elementos de la matriz elevados a potencias de la segunda matriz, por elementos

############################################

def error_predicho_deriv(predicho, target):
    return 2*(predicho- target)

def activacion_sdp_deriv(sdp):
    return sigmoide(sdp)*(1.0-sigmoide(sdp))

def sdp_w_deriv(x):
    return x

def actualiza_w(w, grad, ratio_aprend):
    return w-ratio_aprend*grad
````

````python
# Ingresamos datos
# Definición de entrada, salida, peso y tasa de aprendizaje.
x=0.1
target=0.5
ratio_aprend=0.01
w=np.random.rand()
print('W_inicial:', w)
````

````python
# Implementación paso hacia adelante
z=w*x
predicho=sigmoide(z)
err=error(predicho, target)
print('z:',z)
print('Predicho:',predicho)
print('Error_i:', err)
````

````python
# Implementación paso hacia atras
g1=error_predicho_deriv(predicho, target)
g2 =activacion_sdp_deriv(predicho)
g3=sdp_w_deriv(x)
````

````python
# Calculo del gradiente
grad=g3*g2*g1
print('gradiente',grad)
````

````python
# Calculamos el nuevo peso
w=actualiza_w(w, grad, ratio_aprend)
w
````

El resultado mostrado, es el nuevo peso

**Capas de Keras**

Keras es una API de aprendizaje profundo escrita en Python , que se ejecuta sobre la plataforma de aprendizaje automático Tensorflow . Keras proporciona una gran cantidad de capas preconstruidas para diferentes arquitecturas y propósitos de redes neuronales a través de la API de capas de Keras.

**capa densa**

La capa densa es una red neuronal que tiene una conexión profunda, lo que significa que cada neurona en la capa densa recibe información de todas las neuronas de su capa anterior.

La capa densa realiza una multiplicación matriz-vector, y los valores utilizados en la matriz son parámetros que se pueden entrenar y actualizar con la ayuda de backpropagation.


**Parámetros (Argumentos) de la capa densa de Keras**

**Units:** El parámetro más básico de todos los parámetros, utiliza un entero positivo como valor y representa el tamaño de salida de la capa.

Es el parámetro de la unidad en sí mismo el que juega un papel importante en el tamaño de la matriz de pesos junto con el vector de sesgo .

**activation:** permite aplicar la función de activación por elementos en una capa densa. De forma predeterminada, se usa la activación lineal, pero podemos modificar y cambiar a cualquiera de las muchas opciones que proporciona Keras para esto.

Las siguientes opciones están disponibles como funciones de activación en Keras.

 * Función Relu (activation = relu) – función de activación de la unidad lineal rectificada
 * Función sigmoide(activation = sigmoid) – Función de activación sigmoide, sigmoide(x) = 1 / (1 + exp(-x)).
 * Función Softmax (activation = softmax): softmax convierte un vector de valor en una distribución de probabilidad.
 * Función Softplus(activation = softplus) – Función de activación Softplus, softplus(x) = log(exp(x) + 1).
 * Función Softsign(activation = Softsign) – Función de activación de Softsign, softsign(x) = x / (abs(x) + 1).
 * Función Tanh(activation = tanh) – Función de activación de tangente hiperbólica
 * Función Selu(activation = selu) – Unidad Lineal Exponencial Escalada (SELU).
 * Función Elu(activation = elu) – Unidad Lineal Exponencial
 * Función exponencial (activation = exponential) – Función de activación exponencial.

Revisar: https://www.v7labs.com/blog/neural-networks-activation-functions

**Use_Bias:** El parámetro se usa para decidir si queremos que una capa densa use un vector de sesgo o no. De forma predeterminada, use_bias se establece en verdadero.

**initializers:** El parámetro proporciona los valores de entrada para decidir cómo se inicializarán los valores de capa. Tenemos el vector de sesgo y la matriz de pesos en la capa densa, que debe inicializarse. Revisar: https://keras.io/api/layers/initializers/

 a. kernel_initializer. Este parámetro se utiliza para inicializar la matriz de pesos del kernel. La matriz de pesos es una matriz de pesos que se multiplican con la entrada.

 b. bias_initializer. Un vector de sesgo se puede definir como los conjuntos adicionales de peso que no requieren entrada y corresponden a la capa de salida. Por defecto, se establece como ceros.

**regularizers:** Los regularizadores le permiten aplicar penalizaciones en los parámetros de la capa o en la actividad de la capa durante la optimización. La capa densa tiene 3 regularizadores, kernel_regularizer para la matriz de peso, bias_regularizer para el vector de polarización y activity_regularizer para la salida de la capa. Revisar: https://keras.io/api/layers/regularizers/

  * kernel_regularizer. Representa la función de regularización que se aplicará a la matriz de pesos del kernel.
  * bias_regularizer. representa la función de regularización que se aplicará al vector de bias. Por defecto, está configurado como ninguno.
  * activity_regularizer. Este parámetro se utiliza para la regularización de la función de activación que hemos definido en el parámetro de activación. Se aplica a la salida de la capa. Por defecto, está configurado como ninguno.

**Constraints:** Este último parámetro determina las restricciones sobre los valores que puede tomar la matriz de pesos o el vector de sesgo. Revisar: https://keras.io/api/layers/constraints/

 * kernel_constraint. representa la función de restricción que se aplicará a la matriz de pesos del kernel. . Por defecto, está configurado como ninguno.

* bias_constraint representa la función de restricción que se aplicará al vector de bias. Por defecto, está configurado como ninguno.

**Optimizer**

La optimización es un proceso importante que optimiza los pesos de entrada al comparar la predicción y la función de pérdida. Keras proporciona bastantes optimizadores como módulo:

**Optimizadores disponibles:**

1. **SGD:** Optimizador de descenso de gradiente (con impulso). (Regla de actualización para parámetro w con gradiente g cuando momentum es 0)
2. **RMSprop:** Mantener un promedio móvil (descontado) del cuadrado de los gradientes/Divide el gradiente por la raíz de este promedio.
3. **Adam:** Es un método de descenso de gradiente estocástico que se basa en la estimación adaptativa de momentos de primer y segundo orden.
4. **Adadelta:** Es un método de descenso de gradiente estocástico que se basa en la tasa de aprendizaje adaptativo por dimensión para abordar dos inconvenientes:
  * El decaimiento continuo de las tasas de aprendizaje a lo largo del entrenamiento.
  * La necesidad de una tasa de aprendizaje global seleccionada manualmente.
5. **Adagrad:** Es un optimizador con tasas de aprendizaje específicas de parámetros, que se adaptan en relación con la frecuencia con la que se actualiza un parámetro durante el entrenamiento. Cuantas más actualizaciones recibe un parámetro, más pequeñas son las actualizaciones.
6. **Adamax:** Es una variante de Adán basada en la norma del infinito. Los parámetros predeterminados siguen los proporcionados en el documento. Adamax es a veces superior a adam, especialmente en modelos con incrustaciones.
7. **Nadam:** Optimizador que implementa el algoritmo NAdam. Al igual que Adam es esencialmente RMSprop con impulso, Nadam es Adam con impulso de Nesterov.
8. **Ftrl:** "Follow The Regularized Leader" (FTRL) es un algoritmo de optimización desarrollado en Google para la predicción de la tasa de clics a principios de la década de 2010. Es más adecuado para modelos poco profundos con espacios de funciones grandes y escasos.
